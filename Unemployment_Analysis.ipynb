**Data Visualization**


```python
import plotly.express as px
```

```python
bar=px.bar(ds,x='Region',y=' Estimated Unemployment Rate (%)',color='Region',title='Unemployment Rate',template='plotly')
bar.show()
```

```python
bar2=px.bar(ds,x='Area',y=' Estimated Unemployment Rate (%)',color='Area',title='Unemployment Rate',template='plotly')
bar2.show()
```
